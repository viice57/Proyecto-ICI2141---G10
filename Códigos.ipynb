```python
import numpy as np
import matplotlib.pyplot as plt

# Función que representa la ecuación logística
def ecuacion_logistica(t, y, r, K):
    return r * y * (1 - y / K)

# Función que representa el Método de Runge-Kutta de cuarto orden (RK4), con tasas de crecimiento variables
def runge_kutta(f, t0, y0, h, r, K):
    t = [t0]  # Lista para almacenar los valores de tiempo
    y = [y0]  # Lista para almacenar los valores de población
    año_actual = 0  # Variable para rastrear el año actual

    while t[-1] < T:
        año_actual += 1
        tasa_crecimiento_actual = r[año_actual - 1] / 100  # Obtenemos la tasa de crecimiento actual

        k1 = h * f(t[-1], y[-1], tasa_crecimiento_actual, K)
        k2 = h * f(t[-1] + h/2, y[-1] + k1/2, tasa_crecimiento_actual, K)
        k3 = h * f(t[-1] + h/2, y[-1] + k2/2, tasa_crecimiento_actual, K)
        k4 = h * f(t[-1] + h, y[-1] + k3, tasa_crecimiento_actual, K)

        y_nueva = y[-1] + (k1 + 2*k2 + 2*k3 + k4)/6
        y.append(y_nueva)
        t.append(t[-1] + h)

        print("Tiempo:", t[-1], "años", "Población:", round(y[-1]), "Tasa de Crecimiento:", round(tasa_crecimiento_actual, 3))

    # Redondear la población al entero más cercano después de todas las iteraciones
    y = [round(val) for val in y]
    return t, y

# Parámetros
N0 = 15856977  # Población inicial
K = 18419192  # Capacidad de carga
T = 15  # Intervalo de tiempo en años
dt = 1  # Paso de tiempo en años

# Tasas de crecimiento para cada año (en porcentaje)
tasas_crecimiento = [10.4, 10.0, 10.1, 10.3, 10.8, 10.9, 10.8, 11.1, 10.9, 9.6, 9.9, 10.3, 10.8, 13.8, 17.9]

# Resolver la ecuación logística utilizando el método de Runge-Kutta de cuarto orden
# y representar r como tasas de crecimiento variables
t, y = runge_kutta(ecuacion_logistica, 0, N0, dt, tasas_crecimiento, K)

# Graficar la solución
plt.figure(figsize=(10, 6))
plt.plot(t, y, marker='o', linestyle='-', color = 'b', label='Curva de crecimiento')  # Puntos para cada población
plt.title('Ecuación logística con Tasas de Crecimiento Variables')
plt.xlabel('Tiempo (años)')
plt.ylabel('Población (millones)')
plt.legend()  # Mostrar leyenda
plt.grid(True)
plt.show()
```
